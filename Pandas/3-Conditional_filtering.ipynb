## Filter by a condition:


```python
import numpy as np
import pandas as pd

Data_f = pd.read_csv("tips.csv")
```

```python
Data_f.head()
```

```python
Data_f['total_bill']
```

```python
bool_series = Data_f['total_bill'] > 40
Data_f[bool_series] 
# or we can use: 
# Data_f[Data_f['total_bill'] > 40]
```

```python
Data_f[Data_f['sex'] == "Male"]
```

```python
Data_f[Data_f['size'] > 3]
```

How to work with And & OR

```python
Data_f['total_bill'] > 30
Data_f['sex'] == "Male"
```

For And operator we should use '&'

```python
Data_f[(Data_f['total_bill'] > 30) & (Data_f['sex'] == "Male")]
Data_f.head()
```

For the OR operator we should use '|'

```python
Data_f[(Data_f['day'] == 'Sun') | (Data_f['day'] == 'Friday') | (Data_f['day'] == 'Saturday') ]
```

#### Or we can use this to find and filter - the 'isin' function

```python
options = ['Sat', 'Sun']
Data_f[Data_f['day'].isin(options)]
```
